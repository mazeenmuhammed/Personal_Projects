```python
def create_app():
    external_css = ["https://cdnjs.cloudflare.com/ajax/libs/materialize/0.100.2/css/materialize.min.css"]
    external_js = ['https://cdnjs.cloudflare.com/ajax/libs/materialize/0.100.2/js/materialize.min.js']

    return dash.Dash(external_scripts=external_js, external_stylesheets=external_css)

```

```python
app = dash.Dash(external_stylesheets=[dbc.themes.BOOTSTRAP])
app.layout = html.Div([
    html.H1('Dash Upload Component'),
    dcc.Upload(
        id='upload',
        children=html.Div([
            'Drag and Drop or ',
            html.A('Select Files'),
        ],
        ),style={'width': '93%','height': '60px','lineHeight': '60px','borderWidth': '1px','borderStyle': 'dashed','borderRadius': '5px','textAlign': 'center','margin': '50px'}
    ),
    
    
    html.H4('DataTable'),
    dash_table.DataTable(id='table')
    
],id='container')


@app.callback(
    [Output(component_id='table', component_property='data'), 
    Output(component_id='table', component_property='columns'),
    Output('container', 'style')],
    [Input(component_id="upload",component_property="contents")]
)
def update_figure(contents):
    if contents is None :
        raise PreventUpdate
    else:
        style= {'backgroundColor': '#ff998a'}
        content_string=str(contents).split(',')[1]
        decoded = base64.b64decode(content_string)
        data_frame = create_df(decoded.decode('utf-8'))
        df = data_frame.head(10)
        columns = [{'name': col, 'id': col} for col in df.columns]
        data = df.to_dict(orient='records')
        return data, columns,style
```

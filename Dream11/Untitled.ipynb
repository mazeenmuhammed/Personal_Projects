```python
import requests
from bs4 import BeautifulSoup
# Collect the github page
page = requests.get('https://stats.espncricinfo.com/ci/engine/ground/59396.html?class=3;template=results;type=aggregate')
print(page)
```

```python
soup = BeautifulSoup(page.text, 'html.parser')
print(soup.prettify)
```

```python
soup.find_all('table', class_='engineTable')[3]
```

```python
ls_header = soup.find_all('table', class_='engineTable')[3].find(class_="headlinks").get_text().replace('\n'," ").strip().split()
ls_header
```

```python
ls_of_teams = soup.find_all('table', class_='engineTable')[3].find_all(class_="data1")#[1].get_text().replace('\n'," ").strip()
for i in ls_of_teams:
    print(i.get_text().replace('\n'," ").strip())
```

```python
length = len(ls_of_teams)

ls_dic_header = []
count = 0
while(len(ls_of_teams[count].get_text().replace('\n'," ").strip())>0):
    ls_split = ls_of_teams[count].get_text().replace('\n'," ").strip().split()
    len_header = 0
    dic_header = {}
    if(len(ls_split)==13):
        for header in ls_header:
            dic_header[header] = ls_split[len_header]
            len_header+=1
    else:
        ls_split[0] = ls_split[0]+" "+ls_split[1]
        ls_split.remove(ls_split[1])
        for header in ls_header:
            
            dic_header[header] = ls_split[len_header]
            len_header+=1
    ls_dic_header.append(dic_header)
    count+=1
```

```python
import pandas as pd
df = pd.DataFrame(ls_dic_header).replace("-","")
df = df[df["W/L"]!=""]
df
```

```python
import pandas as pd
from itertools import combinations as cb
from itertools import chain
```

```python
teams ={
    "team1" : 
                {
                    "wk"   : ["dhoni"],
                    "bat"  : ["faf","watson"],
                    "all"  : ["jadeja","sam"],
                    "bowl" : ["ngidi","chahar"]
                },
    "team2" : 
                {
                    "wk"   : ["samson"],
                    "bat"  : ["smith","miller","utappa"],
                    "all"  : [],
                    "bowl" : ["archer","gopal"]
                },
    
}
```

```python
team_df = pd.DataFrame(teams)
team_df
```

```python
wickt_keepers =  teams["team1"]["wk"]+teams["team2"]["wk"]
batsmen = teams["team1"]["bat"]+teams["team2"]["bat"]
bowler = teams["team1"]["bowl"]+teams["team2"]["bowl"]
all_rounder = teams["team1"]["all"]+teams["team2"]["all"]
```

```python
ls = []
for i in range(0,len(wickt_keepers)):
    ls+=list(cb(wickt_keepers,i+1))
wk_comb = ls

ls = []
for i in range(2,len(batsmen)):
    ls+=list(cb(batsmen,i+1))
bat_comb = ls

ls = []
for i in range(2,len(bowler)):
    ls+=list(cb(bowler,i+1))
bowl_comb = ls

ls = []
for i in range(0,len(all_rounder)):
    ls+=list(cb(all_rounder,i+1))
all_comb = ls

tot = wk_comb+bat_comb+bowl_comb+all_comb
ls_tot_comb = list(cb(tot,4))
```

```python
len(wk_comb)+len(bat_comb)+len(bowl_comb)+len(all_comb)
```

```python
def tot_func(x):
    return tuple(set(chain(*x.unique())))

tot_df = pd.DataFrame(ls_tot_comb)
player_df = pd.DataFrame(tot_df.apply(lambda x : tot_func(x), axis=1))
player_df.columns = ["players"]

player_df["len"] = player_df.players.map(lambda x : len(x))
player_df.drop(player_df[player_df.len!=11].index,axis=0, inplace= True)
player_df.drop_duplicates(inplace=True)
player_df.reset_index(inplace=True, drop=True)
player_df
```

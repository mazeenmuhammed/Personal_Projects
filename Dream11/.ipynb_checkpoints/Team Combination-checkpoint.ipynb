```python
teams ={
    "team1" : 
                {
                    "wk"   : ["dhoni"],
                    "bat"  : ["faf","watson"],
                    "all"  : ["jadeja","sam"],
                    "bowl" : ["ngidi","chahar"]
                },
    "team2" : 
                {
                    "wk"   : ["samson"],
                    "bat"  : ["smith","miller","utappa"],
                    "all"  : [],
                    "bowl" : ["archer","gopal"]
                },
    
}
```

```python
import pandas as pd
```

```python
team_df = pd.DataFrame(teams)
team_df
```

```python
from itertools import combinations as cb
from itertools import chain
```

```python
wickt_keepers =  teams["team1"]["wk"]+teams["team2"]["wk"]
batsmen = teams["team1"]["bat"]+teams["team2"]["bat"]
bowler = teams["team1"]["bowl"]+teams["team2"]["bowl"]
all_rounder = teams["team1"]["all"]+teams["team2"]["all"]
```

```python
for i in range(0,len(wickt_keepers)):
    print(list(cb(wickt_keepers,i+1)))
for i in range(2,len(batsmen)):
    print(list(cb(batsmen,i+1)))
for i in range(2,len(bowler)):
    print(list(cb(bowler,i+1)))
for i in range(0,len(all_rounder)):
    print(list(cb(all_rounder,i+1)))
```

```python
ls = []
for i in range(0,len(wickt_keepers)):
    ls+=list(cb(wickt_keepers,i+1))
wk_comb = ls

ls = []
for i in range(2,len(batsmen)):
    ls+=list(cb(batsmen,i+1))
bat_comb = ls

ls = []
for i in range(2,len(bowler)):
    ls+=list(cb(bowler,i+1))
bowl_comb = ls

ls = []
for i in range(0,len(all_rounder)):
    ls+=list(cb(all_rounder,i+1))
all_comb = ls
```

```python
len(wk_comb)+len(bat_comb)+len(bowl_comb)+len(all_comb)
```

```python
df = pd.DataFrame([wk_comb,bat_comb,bowl_comb,all_comb]).T
df.columns = ['wk_comb','bat_comb','bowl_comb','all_comb']
df
```

```python
def func(x, team):
    global teams
    team_mem = list(teams[team].values())
    flat_team_mem = list(chain(*team_mem))
    #print(flat_team_mem)
    count = 0
    for players in x.unique():
        if players!= None:
            for player in players:
                if player in flat_team_mem:
                    #print(player)
                    count+=1
    return count

df2 = df.iloc[:,:3].copy()
df['team1'] = df2.apply(lambda x : func(x,"team1"), axis=1,result_type='broadcast').wk_comb
df['team2'] = df2.apply(lambda x : func(x,"team2"), axis=1,result_type='broadcast').wk_comb
df['total']=df['team1']+df['team2']
df
```

```python
tot = wk_comb+bat_comb+bowl_comb+all_comb
ls_tot_comb = list(cb(tot,4))
```

```python
def func(x, team):
    global teams
    team_mem = list(teams[team].values())
    flat_team_mem = list(chain(*team_mem))
    #print(flat_team_mem)
    count = 0
    for players in x.unique():
        if players!= None:
            for player in players:
                if player in flat_team_mem:
                    #print(player)
                    count+=1
    return count

df3 = df.iloc[:,:3].copy()
df4 = df.iloc[:,:3].copy()
df3['team1'] = df4.apply(lambda x : func(x,"team1"), axis=1,result_type='broadcast').wk_comb
df3['team1'] = df4.apply(lambda x : func(x,"team1"), axis=1,result_type='broadcast').wk_comb
df3['team2'] = df4.apply(lambda x : func(x,"team2"), axis=1,result_type='broadcast').wk_comb
df3['total']=df3['team1']+df3['team2']
df3
```

```python
team1_mem = list(teams["team1"].values())
flat_team1_mem = set(chain(*team1_mem))

ls_team1 = []
ls_team2 = []

def func(x):
    global teams, team1_mem, flat_team1_mem, ls_team1, team2
    #print(flat_team_mem)
    count_t1 = 0
    count_t2 = 0
    team_formed = set(list(chain(*x.unique())))
    #print(team_formed)
    for player in team_formed:
        if player in flat_team1_mem:
            count_t1+=1
        else:
            count_t2+=1
    ls_team1.append(count_t1)
    ls_team2.append(count_t2)

df2 = df.iloc[:,:3].copy()   
    
df2.fillna("", inplace = True)

df2.iloc[:,:3].apply(lambda x : func(x), axis=1)

df2["team1"] = ls_team1
df2["team2"] = ls_team2
df2['total']=df2['team1']+df2['team2']
df2
```

```python
team1_mem = list(teams["team1"].values())
flat_team1_mem = set(chain(*team1_mem))

ls_team1 = []
ls_team2 = []

def func(x):
    global teams, team1_mem, flat_team1_mem, ls_team1, team2
    #print(flat_team_mem)
    count_t1 = 0
    count_t2 = 0
    team_formed = set(chain(*x.unique()))
    #print(team_formed)
    for player in team_formed:
        if player in flat_team1_mem:
            count_t1+=1
        else:
            count_t2+=1
    ls_team1.append(count_t1)
    ls_team2.append(count_t2)

df3 = tot_df.copy()   
    
df3.fillna("", inplace = True)

df3.apply(lambda x : func(x), axis=1)

df3["team1"] = ls_team1
df3["team2"] = ls_team2
df3['total']=df3['team1']+df3['team2']
df3.sort_values("total", ascending = False, inplace=True)
df3[df3.total<12].reset_index(drop=True)
```

```python
def try_func(x):
    return tuple(set(chain(*x.unique())))

try_df = tot_df.copy()
new_try = pd.DataFrame(try_df.apply(lambda x : try_func(x), axis=1))
new_try.columns = ["players"]

new_try["len"] = new_try.players.map(lambda x : len(x))
new_try.drop(new_try[new_try.len!=11].index,axis=0, inplace= True)
new_try.drop_duplicates(inplace=True)
new_try.reset_index(inplace=True, drop=True)
```
